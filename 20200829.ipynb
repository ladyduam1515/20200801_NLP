```python
#-----------------------------PUNTO #1------------------------------
#                 ---leer archivo con with open----

print (data2)
```

```python
#-----------------------------PUNTO #1------------------------------
#                 ---Convertir a minúsculas----


texto = data2.lower() # Minúscula
print(texto)
```

```python
#-----------------------------PUNTO #1------------------------------
#                 ---Conservar sólo palabras----

solo_letras = re.sub(r"[\W\d_]+", " ", texto)
print(solo_letras)
```

```python
#-----------------------------PUNTO #1------------------------------
#                      ---quitar espacios----
sin_espacio = solo_letras.strip()
print(sin_espacio)
```

```python
#-----------------------------PUNTO #1------------------------------
#                        ---Tokenizar----

token = sin_espacio.split()
print(token)
```

```python
#-----------------------------PUNTO #2------------------------------
#                 ---Quitar palabras vacias----

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')
```

```python
sinvacios = [palabra for palabra in token 
         if palabra not in stopwords_sp]
sinvacios
```

```python
#-----------------------------PUNTO #2------------------------------
#    ---¿Cuáles son las 10 palabras no vacias más usadas?----


from collections import Counter

Counter(sinvacios).most_common(10)
```

```python
#-----------------------------PUNTO #3------------------------------
#                ---Stemming del documento----
from nltk.stem.snowball import SnowballStemmer
spanishStemmer = SnowballStemmer ("spanish")

texto_stem =[]

for palabra in sinvacios:
    raiz = spanishStemmer.stem(palabra)
    texto_stem.append(raiz)
print(texto_stem)
```

```python
#-----------------------------PUNTO #3------------------------------
#                ---¿Cuáles son las 10 raíces más usadas?----
from collections import Counter

Counter(texto_stem).most_common(10)
```

```python
#-----------------------------PUNTO #4------------------------------
#---Contar cuántas ocurrencias hay por cada parte de la oración----
####___________________♥¡¡¡¡¡¡¡BONUS!!!!!!!♥________________________


import es_core_news_sm

nlp = es_core_news_sm.load()
doc = nlp( ' '.join(sinvacios))

ocurrencias = []

for w in doc:
    ocurrencias.append(w.pos_)    
    
Counter(ocurrencias).most_common(10)
```
