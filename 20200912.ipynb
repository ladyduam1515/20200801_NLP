```python
import re
import numpy as np
import pandas as pd

from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')

from nltk.stem.snowball import SnowballStemmer
spanishStemmer=SnowballStemmer("spanish")

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

# <font color = "blue">Punto 1: Pre-Procesamiento
- [18 pts]` Leer el archivo `Princesas.csv` usando `pandas` y crear una nueva columna con el texto en minúscula, sin caracteres especiales ni números, sin palabras vacias y hacer stemming de las palabras")</font>

```python
df = pd.read_csv("C:/Users/PC/OneDrive - Grupo Procaps/Documentos/Konrad Lorenz/Semetre_II/NLP/20200912/Princesas.csv")
df
```

```python
#Convierto en minuscula

df["preprocesado"] = df.Personalidad.str.lower()
df
```

```python
#Conservar sólo palabras (letras sin números, sin caractreres especiales)

df["preprocesado"] =  [re.sub(r"[\W\d_]+", " ", solo_pal) for solo_pal in df["preprocesado"]]
df
```

```python
#quitar espacios
df["preprocesado"] =  [solo_letras.strip() for solo_letras in df["preprocesado"]]
df
```

```python
#Stemming del documento
def preprocesado(texto):
    texto = " ".join([spanishStemmer.stem(palabra) for palabra in texto.split() if palabra not in stopwords_sp]) 
    return texto
df['preprocesado'] = df['preprocesado'].apply(lambda texto: preprocesado(texto))
df
```

# <font color= "blue">Punto 2: TF-IDF</font>

-[16 pts]` Crear la matriz TF-IDF

```python
tfidf_vect = TfidfVectorizer()
tfidf = tfidf_vect.fit_transform(df.preprocesado.values)

tfidf_matrix = pd.DataFrame(data=tfidf.toarray(), columns=tfidf_vect.get_feature_names())
tfidf_matrix = tfidf_matrix.T.round(3)
tfidf_matrix.columns = ["Blancanieves","Cenicienta","Aurora","Bella","Jasmín","Pocahontas","Mulan","Tiana","Mérida","Moana"]

#Trasponer
tfidf_matrix
```

# <font color = "blue">Punto 3: Distancia del coseno</font>
[12 pts]` Calcular la distancia del coseno entre cada una de las princesas

```python
from sklearn.metrics.pairwise import cosine_distances

dist_cos = cosine_distances(tfidf_matrix.T.values)
dist_cos = pd.DataFrame(dist_cos, columns = tfidf_matrix.columns, index = tfidf_matrix.columns)

dist_cos
```

# <font color = "blue"> [2 pts] ¿Cuáles son las princesas más diferentes? </font>

```python
listado = ["Blancanieves","Cenicienta","Aurora","Bella","Jasmín","Pocahontas","Mulan","Tiana","Mérida","Moana"]
for nombre in listado:
    result = dist_cos [nombre].idxmax()
    print(result +" es la más diferente de " + nombre)
```
